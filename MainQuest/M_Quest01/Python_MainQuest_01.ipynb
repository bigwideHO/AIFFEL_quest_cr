```python
# Q10

import random

class Account():
    account_count = 0

    def __init__(self, holder, balance):
        self.holder = holder
        self.balance = balance

        self.bank_name = "SC은행"

        self.deposit_count = 0
        self.withdraw_count = 0
        self.deposit_log = []
        self.withdraw_log = []
        
        num01 = random.randint(0, 999)
        num02 = random.randint(0, 99)
        num03 = random.randint(0, 999999)

        num01 = str(num01).zfill(3)
        num02 = str(num02).zfill(2)
        num03 = str(num03).zfill(6)

        self.account_num = num01 + '-' + num02 + '-' + num03

        Account.account_count += 1

    @classmethod
    def get_account_num(cls):
            print(cls.account_count)
    
    def deposit(self, amount):
        if amount >= 1:
            self.deposit_count += 1
            self.de_type = "입금"
            self.deposit_log.append(amount)
            self.balance += amount
            self.de_amount = amount
        else:
            print("입금은 최소 1원 이상만 가능합니다.")

        if self.deposit_count % 5 == 0:
            self.in_type = "이자 지급"
            self.just_before_balance = self.balance
            self.balance = (self.balance * 1.01)
            self.interest = self.balance - self.just_before_balance
            self.deposit_count += 1
            print(f"{self.interest}원의 이자가 입금되었습니다.")
    
    def withdraw(self, amount):
        if amount < self.balance:
            self.withdraw_count += 1
            self.wi_type = "출금"
            self.withdraw_log.append(amount)
            self.balance -= amount
            self.wi_amount = amount
        else:
            print("계좌 잔고 이상의 금액으로 출금할 수 없습니다.")
    
    def display_info(self):
        print(f"은행 이름: {self.bank_name}, 예금주: {self.holder}, 계좌번호: {self.account_num}, 잔고: {self.balance:,}")
    
    def withdraw_history(self):
        for self.withdraw_count in self.withdraw_log:
            print(f"{withdraw_count}회: {self.wi_type}\t금액: {self.de_amount}\t잔액: {self.balance}")

    def deposit_history(self):
        for self.deposit_count in self.deposit_log:
            if self.de_type == "입금":
                print(f"{self.deposit_count}회: {self.de_type}\t금액: {self.wi_amount}\t잔액: {self.balance}")
            else:
                print(f"{self.deposit_count}회: {self.in_type}\t금액: {self.interest}\t잔액: {self.balance}")

my_account = Account('차정은', 1000)

for i in range(10):
    my_account.deposit(1000)
my_account.deposit(-1000)
my_account.withdraw(3000000)
my_account.withdraw(3000)

my_account.deposit_history()
print('-' * 50)
my_account.withdraw_history()


```
